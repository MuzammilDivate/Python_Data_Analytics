# pie charts

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ast

df = pd.read_csv("C:/Users/DELL/Downloads/data_jobs.csv")
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.head()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', autopct='%0.1f%%',startangle=90, title='Work from home status')
plt.ylabel('')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='pie', autopct='%0.1f%%', title='Percent of jobs')
plt.ylabel('')
plt.show()
```

```python
df.columns
```

```python
job_titles = {
    'job_work_from_home':'Work from home status',
    'job_no_degree_mention':'Job degree req.',
    'job_health_insurance':'Health insurance offered'
}
```

```python
fig, ax = plt.subplots(1, 3)
for i, (col, tit) in enumerate(job_titles.items()):
    # print(i, col, tit)
    ax[i].pie(df[col].value_counts(), startangle=90, autopct='%0.1f%%', labels=['False', 'True'])
    ax[i].set_title(tit)

plt.show()
```
