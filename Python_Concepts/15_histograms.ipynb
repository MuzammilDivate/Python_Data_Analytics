```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ast

df = pd.read_csv("C:/Users/DELL/Downloads/data_jobs.csv")
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_da_ind = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='India')].copy()
```

```python
df_da_ind.head()
```

```python
df_da_ind['salary_year_avg'].plot(kind='hist', bins=50, edgecolor='black')
plt.xlim(0,200000)
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
plt.xlabel("Yearly Salary")
plt.ylabel('No. of jobs')
plt.title('Distribution of Indian Data Analyst Yearly Salaries')
plt.show()
```

```python
# ill make one for us as well

df_us_da = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')].copy()
```

```python
df_us_da.head()
```

```python
df_us_da['salary_year_avg'].plot(kind='hist', bins=40, edgecolor='k')
plt.xlim(0, 280000)
ax=plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _:f'${int(x/1000)}K'))
plt.xlabel("Yearly Salary")
plt.ylabel('No. of jobs')
plt.title('Distribution of US Data Analyst Yearly Salaries')
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10,8))

df_us_da['salary_year_avg'].plot(kind='hist', bins=40, edgecolor='k', ax=axs[0])
axs[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _:f'${int(x/1000)}K'))
axs[0].axvline(df_us_da['salary_year_avg'].median(), color='r', linestyle='--')
axs[0].axvline(df_us_da['salary_year_avg'].quantile(0.25) ,color='y', linestyle='--')
axs[0].axvline(df_us_da['salary_year_avg'].quantile(0.75), color='y', linestyle='--')
axs[0].legend(['','Median', 'Q1 & Q3'])
axs[0].set_xlabel("Yearly Salary")
axs[0].set_ylabel('No. of jobs')
axs[0].set_title('Distribution of US Data Analyst Yearly Salaries')

df_us_da['salary_year_avg'].plot(kind='box', ax=axs[1], vert=False)
axs[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _:f'${int(x/1000)}K'))
axs[1].axvline(df_us_da['salary_year_avg'].median(), color='r', linestyle='--')
axs[1].axvline(df_us_da['salary_year_avg'].quantile(0.25), color='y', linestyle='--') 
axs[1].axvline(df_us_da['salary_year_avg'].quantile(0.75), color='y', linestyle='--') 
axs[1].legend(['', 'Median', 'Q1 & Q3'])
axs[1].set_xlabel('Median Salary($USD)') 
axs[1].set_ylabel('') 
plt.tight_layout()

plt.show()
```

```python
# now lets  plot 3 boxex on one plot
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_ind = df[(df['job_title_short'].isin(job_titles)) & (df['job_country']=='India')].copy()
```

```python
df_ind = df_ind.dropna(subset=['salary_year_avg'])
```

```python
job_lists = [df_ind[df_ind['job_title_short']==jobtitle]['salary_year_avg'] for jobtitle in job_titles]
```

```python
job_lists[0]
```

```python
plt.boxplot(job_lists, tick_labels=job_titles, vert=False)
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _:f'${int(x/1000)}K'))
plt.xlabel('Salary ($USD)')
plt.title('Salary distribution in India')
plt.xlim(0, 350000)

plt.show()
```

# Now for US using the same code


```python
df_US = df[(df['job_title_short'].isin(job_titles)) & (df['job_country']=='United States')].copy()
df_US = df_US.dropna(subset=['salary_year_avg'])
job_list2 = [df_US[df_US['job_title_short']==jobtitle]['salary_year_avg'] for jobtitle in job_titles]
```

```python
plt.boxplot(job_list2, tick_labels=job_titles, vert=False)

ax = plt.gca()

ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _:f'${int(x/1000)}K'))

plt.xlabel('Salary ($USD)')
plt.title('Salary distribution in the United States')
plt.xlim(0, 600000)
plt.show()
```
