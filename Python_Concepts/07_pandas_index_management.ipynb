# Index Management
* .reset_index()
* .sort_index()
* .set_index()

```python
# Lets import all the necessary stuff

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("C:/Users/DELL/Downloads/data_jobs.csv")
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df.head()
```

```python
df.index

# Currently the index of the df is a range from 1 to len(df)
```

# .set_index()

sets a column as the index of the df

```python
# We can choose any column to be the new index for our df

df_set_index = df.set_index('job_title_short') # -> now 'job_title_short will be our new index 

df_set_index # -> Yup
```

# .reset_index()
 
resets the index of the df to a numeric array like 0, 1, 2, 3, 4 and so on

```python
# Now lets say we want our data to have the original numeric index we can use the reset_index()

df_reset_index = df_set_index.reset_index() #-> Or i can also use inplace=True

df_reset_index.head() #-> Yup
```

# .sort_index()

Sorts the df in order to the index

Now lets take the df_set_index df again once and well sort the values in the df in an ordered manner based on the index

```python
df_sorted = df_set_index.sort_index()

df_sorted.head() #-> Done
```
