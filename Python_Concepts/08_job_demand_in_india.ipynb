```python
# Lets import all the necessary stuff

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("C:/Users/DELL/Downloads/data_jobs.csv")
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df.head()
```

```python
df_ind = df[df['job_country']=='India'].copy()
```

```python
df_ind['job_month'] = df_ind['job_posted_date'].dt.strftime('%B')  #-- '%B' for the name of the month
df_ind.head()
```

```python
# Now running the pivot_table method on this

df_ind_pivot = df_ind.pivot_table(index='job_month', columns='job_title_short', aggfunc='size')

# Now as you can see the index is in an aplphabetical order but we want it in a chronological order like jan>feb>mar so on
```

```python
df_ind_pivot
```

```python
# First lets reset the index
df_ind_pivot.reset_index(inplace=True)
```

```python
df_ind_pivot
```

```python
# So to solve that we can do a reverse of the pd.to_datetime()

df_ind_pivot['job_mon_'] = pd.to_datetime(df_ind_pivot['job_month'], format='%B').dt.month 
```

```python
df_ind_pivot.sort_values('job_mon_', inplace=True)
```

```python
df_ind_pivot.set_index('job_month', inplace=True)
```

```python
df_ind_pivot.drop(columns='job_mon_', inplace=True)
```

```python
job_titles = df['job_title_short'].value_counts().head(3)
job_titles = job_titles.index.tolist()
job_titles
```

```python
df_ind_pivot = df_ind_pivot[job_titles]
```

```python
df_ind_pivot
```

```python
df_ind_pivot.plot(kind='line', 
                  linewidth=4,
                  linestyle='--',
                  colormap='inferno',
                  marker='o',
                  markersize=6,
                  figsize=(8,6)
                  )
plt.grid()
plt.xlabel('Month (2023)')
plt.ylabel('Count')
plt.title('Median salary of jobs by Months')
```
